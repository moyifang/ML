```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import numpy as np

# Generating random data
np.random.seed = 10
size = 200
X = np.random.random((size, 2))

# Applying transformation
transformation = [[0.60834549, -0.63667341], [-0.40887718, 0.85253229]]
X = np.dot(X, transformation)

# Performing PCA transformation
pca = PCA(n_components=2).fit(X)
X_transformed = pca.transform(X)

# Plotting
plt.rcParams.update({'font.size': 10})
plt.figure(figsize=(13, 6))
plt.subplot(121)
plt.scatter(X[:, 0], X[:, 1], color='red')
plt.xlim([-1, 1])
plt.ylim([-1, 1])
plt.title('Raw data')
plt.subplot(122)
plt.xlim([-1, 1])
plt.ylim([-1, 1])
plt.title('PCA rotation ')
plt.scatter(X_transformed[:, 0], X_transformed[:, 1], color='blue')
plt.suptitle("Explained variance ratio (all components): %s" % str(pca.explained_variance_ratio_))
plt.savefig('PCA rotation.png', dpi=600)
plt.show()
print("explained variance ratio (all components): %s" % str(pca.explained_variance_ratio_))
```
